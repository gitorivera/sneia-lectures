# Preprocesamiento de los datos
Con el fin de brindarle a la red neuronal los datos en un formato que sea fácil de manejar se realizarán por la red neuronal. Los pasos del preprocesamieto son los siguientes:


1.   Se transforman los datos de entrada de dos dimensiones (imágenes), a una dimensión (vectores). Este proceso es necesario ya que las redes neuronales trabajan con vectores.
2.   Se normalizan los valores entre 1 y 0, esto se logra dividiendo por 255 el cuál es el máximo valor que tiene un píxel de la imagen (color blanco).
3. Se codifican las etiquetas en forma de one-hot encoding. El one hot encoding es una técnica de representación de variables categóricas en forma numérica que se utiliza frecuentemente en aprendizaje automático y redes neuronales.

Consiste en transformar cada valor categórico en un vector binario que tenga un único valor 1 en la posición correspondiente al valor, y 0 en el resto.

Por ejemplo, si tenemos 3 valores posibles para una variable: [A, B, C]:
A se convierte en [1, 0, 0]
B se convierte en [0, 1, 0]
C se convierte en [0, 0, 1]
De esta manera se obtiene una representación numérica que permite utilizar algoritmos matemáticos preservando la naturaleza categórica de los datos.

Las ventajas de one hot encoding son:

Permite representar valores discretos de forma numérica
No implica orden entre los valores (A no es menor que B)
Evita asignar valores numéricos arbitrarios a las categorías
Permite aplicar algoritmos que requieren datos numéricos
