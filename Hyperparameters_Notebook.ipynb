# Preprocesamiento de los datos
Con el fin de brindarle a la red neuronal los datos en un formato que sea fcil de manejar se realizarn por la red neuronal. Los pasos del preprocesamieto son los siguientes:


1.   Se transforman los datos de entrada de dos dimensiones (imgenes), a una dimensin (vectores). Este proceso es necesario ya que las redes neuronales trabajan con vectores.
2.   Se normalizan los valores entre 1 y 0, esto se logra dividiendo por 255 el cul es el mximo valor que tiene un pxel de la imagen (color blanco).
3. Se codifican las etiquetas en forma de one-hot encoding. El one hot encoding es una tcnica de representacin de variables categricas en forma numrica que se utiliza frecuentemente en aprendizaje automtico y redes neuronales.

Consiste en transformar cada valor categrico en un vector binario que tenga un nico valor 1 en la posicin correspondiente al valor, y 0 en el resto.

Por ejemplo, si tenemos 3 valores posibles para una variable: [A, B, C]:
A se convierte en [1, 0, 0]
B se convierte en [0, 1, 0]
C se convierte en [0, 0, 1]
De esta manera se obtiene una representacin numrica que permite utilizar algoritmos matemticos preservando la naturaleza categrica de los datos.

Las ventajas de one hot encoding son:

Permite representar valores discretos de forma numrica
No implica orden entre los valores (A no es menor que B)
Evita asignar valores numricos arbitrarios a las categoras
Permite aplicar algoritmos que requieren datos numricos
